```python
#Installed required libraries
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Read dataset
df = pd.read_csv(r"res_iter.csv", header = None)
df
```

```python
# Fix formatting
df = df.iloc[:,:3]
df.columns = ['Azimuth','Elevation','Entropy']
```

```python
x = df['Azimuth']
y = df['Elevation']
z = df['Entropy']

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
#ax.plot3D(x, y, z)
ax.scatter3D(x, y, z, c=z, cmap='Reds')


# Set labels and title
ax.set_xlabel('Azimuth')
ax.set_ylabel('Elevation')
ax.set_zlabel('Entropy')
ax.set_title('Optimisation Process Visualisation')
plt.savefig("optimisation_process_visualisation")
# Show the plot
plt.show()
```
